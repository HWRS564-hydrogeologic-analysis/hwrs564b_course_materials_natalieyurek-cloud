```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erfc

# Column parameters
L = 0.5          # m
V = 0.5          # m/day
Pe_values = [5, 50, 500]

# Time array
t = np.linspace(0.001, 5, 2000)

plt.figure(figsize=(8,6))

for Pe in Pe_values:
    D = V * L / Pe
    
    C = 0.5 * erfc((L - V*t) / (2*np.sqrt(D*t))) + \
        0.5 * erfc((L + V*t) / (2*np.sqrt(D*t)))
    
    plt.plot(t, C, label=f'Pe = {Pe}')

# Formatting
plt.xlabel('Time (days)')
plt.ylabel('C(L,t)/C0')
plt.title('Breakthrough Curves at Column Outlet')
plt.legend()
plt.grid(True)
plt.ylim(0,1.05)

plt.show()

```
